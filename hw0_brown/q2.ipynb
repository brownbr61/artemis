## 2c)
Get the angle θ between ~c and ~a. Get the unit vector of ~c in two ways (by shrinking the vector; by using the angle θ). Then create vectors that has the same length as ~c but with the angle that is 3,4 times of θ.

```python
# angle = arccos[(xa * xb + ya * yb) / (√(xa2 + ya2) * √(xb2 + yb2))]
a = np.array([1,0])
b = np.array([0,1])
c = a + b
theta = np.arccos(a@c/np.linalg.norm(c))
print("Angle between leg and hypotenuse (theta): \t\t{0} radians".format(theta))

unitc = c/np.linalg.norm(c)
print("unit vector for hypotenuse by shrinking (unitc): \t{}".format(unitc))
unitc = np.array([np.cos(theta),np.sin(theta)])
print("unit vector for hypotenuse from the angle (unitc): \t{}\n".format(unitc))

c3 = np.array([np.cos(3*theta),np.sin(3*theta)])
c4 = np.array([np.cos(4*theta),np.sin(4*theta)])
print("unit vector for hypotenuse from 3*theta (c3): \t\t{}".format(c3))
print("unit vector for hypotenuse from 4*theta (c4): \t\t{}".format(c4))
```

Plot all the vectors we have created. Check the documentation for plot to see how

```python
v = np.array([a,b,c,unitc,c3,c4])

for i in np.arange(len(v[:,1])):
    plt.plot([0,v[i,0]],[0,v[i,1]],'-o')

plt.xlabel("x component")
plt.ylabel("y component")
```
